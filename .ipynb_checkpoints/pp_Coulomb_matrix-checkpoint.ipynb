```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
#initialization of the coulom matrix. 
H1 = np.zeros((15,15))
H1
```

```python
Diagonal_coeffecient_list = [1, -5, -2, -2,-5, -5, 1, 1, 4, -5, -5, -2,-2, -5, 1]
Diagonal_coeffecient_list = [ 1/25 * x for x in Diagonal_coeffecient_list]
Off_diagonal_elements = [-6/25,-3/25, 3/25]
```

```python
for i in range(H1.shape[0]):
    for j in range(H1.shape[1]):
        if i == j:
            H1[i,j] =  Diagonal_coeffecient_list[i]
        if ((i == 2) & (j == 3 )) or ((i == 8) & (j == 7)):
            H1[i,j] = Off_diagonal_elements[1] 
            H1[j,i] = H1[i,j]
        if ((i == 6) & (j == 7)) or ((i == 11) &(j == 12)):
            H1[i,j] = Off_diagonal_elements[0] 
            H1[j,i] = H1[i,j]   
        if ((i == 8) & (j == 6)):
            H1[i,j] = Off_diagonal_elements[2] 
            H1[j,i] = H1[i,j]        
df = pd.DataFrame(H1)
```

```python
df
```

```python
eigenvalues , eigensvec = np.linalg.eig(H1)
```

```python
D = np.diag(eigenvalues)
```

```python
p = eigensvec
p_inv =np.linalg.inv(p)
```

```python
pd.DataFrame(p*D*p_inv)
```
