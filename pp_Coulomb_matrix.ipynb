```python
import numpy as np
import sympy  as sp
import matplotlib.pyplot as plt
```

```python
#initialization of the coulom matrix. 
F0 = 114.848732#00516
F2 = 54.126484#473545

Diagonal_coeffecient_list_F2 = [1, -5, -2, -2,-5, -5, 1, 1, 4, -5, -5, -2,-2, -5, 1]
Diagonal_coeffecient_list_F0_F2 = [ (F0+ (F2*1/25 * x)) for x in Diagonal_coeffecient_list_F2]
H1 = sp.zeros(15,15)
H1
```

```python
Off_diagonal_elements_F2 = [-6/25,-3/25, 3/25]
Off_diagonal_elements_F2=[F2*x for x in Off_diagonal_elements_F2]
```

```python
for i in range(H1.shape[0]):
    for j in range(H1.shape[1]):
        if i == j:
            H1[i,j] =  Diagonal_coeffecient_list_F0_F2[i]
        if ((i == 2) & (j == 3 )) or ((i == 8) & (j == 7)):
            H1[i,j] = Off_diagonal_elements_F2[1] 
            H1[j,i] = H1[i,j]
        if ((i == 6) & (j == 7)) or ((i == 11) &(j == 12)):
            H1[i,j] = Off_diagonal_elements_F2[0] 
            H1[j,i] = H1[i,j]   
        if ((i == 8) & (j == 6)):
            H1[i,j] = Off_diagonal_elements_F2[2] 
            H1[j,i] = H1[i,j]        
H1
```

```python
P, D = H1.diagonalize()
```
